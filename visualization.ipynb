```
# To keep the page organized do all imports here
import pandas as pd
import seaborn as sns 
sns.set()

# suppress warnings (warnings do not prevent the code from executing)
import warnings
warnings.filterwarnings("ignore")
```

```
hp_path = '/content/drive/My Drive/Thinkful DS (trial) /Data/houseprices.csv'
```

```
houseprices = pd.read_csv(hp_path)
houseprices.info()
houseprices.head()
```

```
# variation in saleprices
v_sp = sns.distplot(houseprices['saleprice']/1000, 
                    bins = 50,
                    axlabel = "Saleprice (Th $)")

# tried rotating axislabels but to no avail. Hence converted axis into thousands
```

```
# variation in lotarea 
v_la = sns.distplot(houseprices['lotarea'], 
                    bins = 50,
                    axlabel = 'Lot area (sq.ft)')
# most are small plots (as expected) between 100 and 20000 sq.ft
```

```
# boxplot showing variation in saleprices by overall condition 
v_sp_oc = sns.boxplot(y="saleprice", x="overallcond", data=houseprices)
```

```
# boxplot showing variation in saleprices by overall condition with and without central a/c
v_sp_oc_ca = sns.boxplot(y="saleprice", x="overallcond", hue='centralair', data=houseprices)
```

```
# scatterplot to examin relationship between lotarea and saleprice
sns.jointplot(x="lotarea", y="saleprice", data=houseprices)

# seems to be a weak correlation visually and confirmed by the small value (<0.3) 
houseprices['lotarea'].corr(houseprices['saleprice'])
```
