Applying Central Limit Theorem and the Law of Large Numbers to Binomial and Posson distributions 

```
# Initialize environment

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from numpy.random import randint
from numpy import mean
from scipy import stats
from numpy.random import seed
import seaborn as sns 
sns.set()

# suppress warnings (warnings do not prevent the code from executing)
import warnings
warnings.filterwarnings("ignore")
```

### Line charts plotting the sample mean for Binomial and Poisson distributions as more and more trials are conducted, up to 1,000 trials.

#### Binomial

```
results = []
for i in range(1, 1000):
  # sample from binomial distribution. pick any digit upto 99 and any probability p
  sample = np.random.binomial(25, 0.25, i) # random.randint(1, 99), 0.25, i)
  # calculate mean
  mean_of_trials = sample.mean()
  # add mean
  results.append(mean_of_trials)

# create a df
df = pd.DataFrame({"results": results})

# plot mean
df.plot()

# Binomial
# if you randomly vary n and p, the plot is horrendous
# trying random n with same p: still not much change
# trying same n and p: works!
```

#### Poisson

```
results = []
for i in range(1, 1000):
  # sample from poisson distribution. pick any digit upto 99 and any probability p
  sample = np.random.poisson(10, i) # random.randint(1, 99), 0.25, i)
  # calculate mean
  mean_of_trials = sample.mean()
  # add mean
  results.append(mean_of_trials)

# create a df
df = pd.DataFrame({"results": results})

# plot mean
df.plot()

# Poisson
# if you randomly vary lambda, the plot is horrendous which makes sense
# cuz you are asking it to oick any value whose mean will be anything and there's 
# no way it's gonna stabilize as a demo of LoLN 
# trying fixed lambda: works!
```

### Randomly sampling the mean 1,000 times and plotting the resulting frequency of sample means as a histogram for Binomial and Poisson distributions

#### Binomial

```

# set seed to get conistent and verifiable results
seed(1234)
means = [np.random.binomial(25, 0.25, 100).mean() for i in range(1000)]

# plot histogram
plt.hist(means)
plt.show()

print(stats.describe(means))
```

#### Poisson

```
# Poisson
# set seed to get conistent and verifiable results
seed(1234)
means = [np.random.poisson(10, 100).mean() for i in range(1000)]

# plot histogram
plt.hist(means)
plt.show()

print(stats.describe(means))
```
