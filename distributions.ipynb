Plotting histograms for different random distributions

```
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns 
sns.set()

# suppress warnings (warnings do not prevent the code from executing)
import warnings
warnings.filterwarnings("ignore")
```

```
# choose 6 distributions that take exactly 3 arguments, with the last one being size
distributions = ['gumbel', 'normal', 'laplace', 'logistic', 'lognormal', 'uniform']

'''
  takes 3 arguments needed to define a six kinds of distribution and plots a 
  histogram for each distribution with mean +/- 1 std deviations displayed
'''

def draw_hists(a1, a2, size):
  for i in distributions:
    # generate a df for each distribution from the list of 6 distr. 
    df =  pd.DataFrame({i: getattr(np.random, i)(a1, a2, size)})
    
    # Plot a histogram
    df.hist(color ='gold', bins = 20)

    # Add a vertical line at the mean.
    plt.axvline(df.mean().values, color='b', 
                linestyle='solid', linewidth=2)

    # Add a vertical line at one standard deviation above the mean.
    plt.axvline(df.mean().values + df.std().values, color='b', 
                linestyle='dashed', linewidth=2)

    # Add a vertical line at one standard deviation below the mean.
    plt.axvline(df.mean().values - df.std().values, color='b', 
                linestyle='dashed', linewidth=2) 

    # Print the histogram.
    plt.show()
```

```
draw_hists(random.randint(1,10), random.randint(1,10), 100)

"""
  Except for lognormal distribution and to an extent uniform, mean and SD give a good sense of the 
  spread of distributions.  Judging visually, logistic and lognormal distributions have more data within 
  1 SD of the mean as compared to the 68% in Normal, Gumbel, Laplace and Uniform distributions.

"""
```

```
# Making two variables.
rand1 = np.random.normal(5, 0.5, 1000)
rand2 = np.random.normal(10, 1, 1000)

# summing rand1 and rand2 to get another variable
rand3 = rand1 + rand2

# generate a df for each distribution from the list of 6 distr. 
df =  pd.DataFrame({'rand': rand3})

# Plot a histogram
df.hist(color ='saddlebrown', bins = 20)

# Add a vertical line at the mean.
plt.axvline(df.mean().values, color='b', 
            linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(df.mean().values + df.std().values, color='b', 
            linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(df.mean().values - df.std().values, color='b', 
            linestyle='dashed', linewidth=2) 

# Print the histogram.
plt.show()

'''
  Sum of 2 normal distributions is also normal. The mean (15) +/- 1SD (1.1) 
  where mean is the sum of 2 random distributions and the SD being the square 
  root of the sum of the square of constituent SDs i.e. sqrt(1^2 + 0.5^2)
'''
```
